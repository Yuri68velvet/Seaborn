# distplot
* The distplot shows the distribution of a univeriate set of observations.

```python
sns.distplot(diamonds['carat'])
```

#### To remove the KDE layer and just have the histgram use:

```python
sns.distplot(diamonds['carat'],kde=False)
```

# jointplot

#### jointplot() allows you to basically match up two distplots for bivarate data.With your choice of what kind parameter to compare with:

* "scatter"
* "reg"
* "resid"
* "kde"
* "hex"

```python
#sns.jointplot(x='totalbill',y='tip',data='tips',kind='scatter')
sns.jointplot(x='carat',y='price',data=diamonds,kind='scatter')
```

```python
#sns.jointplot(x='total_bill',y='tip',data=diamonds,kind='hex')
sns.jointplot(x='carat',y='price',data=diamonds,kind='hex')
```

# pair plot
pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue aargument (for categorical columns).

```python
sns.pairplot(diamonds)
```

```python
sns.pairplot(diamonds,hue='color',kind='scatter')
```

# rugplot
rugplot are actually a very simple concept,they just draw a dash mark for every point on a univariate distribution.They are the building block of a KDE plot:

```python
sns.rugplot(diamonds['carat'])
```
