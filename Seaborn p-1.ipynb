```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

# DistPlot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

# JointPlot

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='scatter')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

# Pairplot

```python
sns.pairplot(tips)
```

# Hue

```python
sns.pairplot(tips,hue='sex',kind='scatter')
```

# Rugplot

```python
sns.rugplot(tips['total_bill'])
```

# Rugplot+KDEplot

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# Barplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.sum)
```

# Countplot

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(data=tips,palette='pastel',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='coolwarm')
```

# Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='pastel')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True,palette='Set1')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,palette='Set1',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips,palette='Set1',split=True)
```

```python
sns.violinplot(x='total_bill',y='day',data=tips,palette='pastel')
sns.swarmplot(x='total_bill',y='day',data=tips,color='black',size=3)
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```

# Matrixplot

```python
flight=sns.load_dataset('flights')
```

```python
flight.head()
```

# Heatmap

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(flight.corr())
```

```python
sns.heatmap(tips.corr(),cmap='Set1',annot=True)
```

```python
flight_data=flight.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flight_data)
```

```python
sns.heatmap(flight_data,cmap='coolwarm',linecolor='grey',linewidth=0.5)
```

# Clustermap

```python
sns.clustermap(flight_data)
```

```python
sns.clustermap(flight_data,cmap='coolwarm')
```
